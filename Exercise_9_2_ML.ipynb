```python
%%capture
import numpy as np
from nilearn import datasets
from nilearn.input_data import NiftiLabelsMasker
from nilearn.image import index_img
import nibabel as nib
```

```python
%%capture
import matplotlib.pyplot as plt
from numpy import mean
from numpy import std
from sklearn.linear_model import LogisticRegression
from sklearn.utils import resample
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
import pandas as pd
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import KFold
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from sklearn.decomposition import FastICA
from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV
```

```python
%%capture
from IPython import display
display.set_matplotlib_formats('svg')
```

```python
# We import the data first in order to select the last 10 subjects
%%capture
brain_data = datasets.fetch_oasis_vbm(n_subjects=(100))
yeo = datasets.fetch_atlas_schaefer_2018(n_rois=(100)) # We fix this at 100
masker = NiftiLabelsMasker(labels_img=yeo.maps, standardize=True, memory='nilearn_cache')
input_variables = masker.fit_transform(brain_data.gray_matter_maps)
output_variable = np.array(brain_data.ext_vars.mf == 'F', dtype=int) # gives 1 for females and 0 for males
```

```python
# We prepare the X matrix and y vector, and we scale X
X, y = input_variables, output_variable
X_scaled = scaler.fit_transform(X)
```

```python
# We create the train-test split by hand
train_indices = list(range(0,90))
test_indices = list(range(90,100))
# We combine both splits into one variable
split = [(train_indices, test_indices)] 
```

```python
%%capture
# We prepare the MLPRegressor model with the parameters we want fixed
mlp = MLPClassifier(max_iter=1000, random_state=42)
# We create a variable containing the parameter space we want to explore
parameters = {'alpha':[0.0001, 0.001, 0.01, 0.1, 1.0], 'hidden_layer_sizes':[[5,], [10,], [25,], [50,], [100,]]}
# We call the GridSearch estimator
clf = GridSearchCV(mlp,parameters,cv=split,return_train_score=True)
# We fit the GridSearch estimator on our data
clf.fit(X_scaled, y)
```

```python
# We store the results into a pandas dataframe
NN = pd.DataFrame(clf.cv_results_)
```

```python
# We retrieve the two columns of interest 
test_scores = np.array(NN["split0_test_score"])
train_scores = np.array(NN["split0_train_score"])
```

```python
# We divide the data according to layer depth
a = test_scores[0:5]
b = test_scores[5:10]
c = test_scores[10:15]
d = test_scores[15:20]
e = test_scores[20:25]
```

```python
# We divide the data according to layer depth
f = train_scores[0:5]
g = train_scores[5:10]
h = train_scores[10:15]
i = train_scores[15:20]
j = train_scores[20:25]
```

```python
# We regroup the results
test_test_scores = [a,b,c,d,e]
train_train_scores = [f,g,h,i,j]
```

```python
# We're ready to plot the training scores!
width = list(range(1,6))
lab = ['0.0001', '0,001', '0.01', '0.1', '1.0']
plt.style.use("default")
plt.scatter(width,train_train_scores[0], s=25, alpha = 0.3, marker = 'o')
plt.scatter(width,train_train_scores[1], s=25, alpha = 0.3, marker = 'o')
plt.scatter(width,train_train_scores[2], s=25, alpha = 0.3, marker = 'o')
plt.scatter(width,train_train_scores[3], s=25, alpha = 0.3, marker = 'o')
plt.scatter(width,train_train_scores[4], s=25, alpha = 0.3, marker = 'o')
plt.legend(labels=(lab))
plt.gca().set(
    title="MLP CLASSIFIER 2: Training set model accuracy as a function of layer width (5-100) and alpha values (0.0001-1.0) for 100 ROIs",
    xlabel="Layer width (5-100)",
    ylabel="Classification accuracy (0-1) for the training sets",
    xticks = (width),
    xticklabels = [5,10,25,50,100]
)
plt.plot(width,train_train_scores[0])
plt.plot(width,train_train_scores[1])
plt.plot(width,train_train_scores[2])
plt.plot(width,train_train_scores[3])
plt.plot(width,train_train_scores[4])
plt.show()
```

A minimum of 10 hidden units is required to achieve a perfect training accuracy across all 5 values of alpha.

```python
# We're ready to plot the test scores!
width = list(range(1,6))
lab = ['0.0001', '0,001', '0.01', '0.1', '1.0']
plt.style.use("default")
plt.scatter(width,test_test_scores[0], s=50, alpha = 0.5, marker = 'o')
plt.scatter(width,test_test_scores[1], s=50, alpha = 0.5, marker = 'd')
plt.scatter(width,test_test_scores[2], s=50, alpha = 0.5, marker = 'p')
plt.scatter(width,test_test_scores[3], s=50, alpha = 0.5, marker = 'h')
plt.scatter(width,test_test_scores[4], s=50, alpha = 0.5, marker = 'h')
plt.legend(labels=(lab))
plt.gca().set(
    title="MLP CLASSIFIER 2: Test set model accuracy as a function of layer width (5-100) and alpha values (0.0001-1.0) for 100 ROIs",
    xlabel="Layer width (5-100)",
    ylabel="Classification accuracy (0-1) for the test sets",
    xticks = (width),
    xticklabels = [5,10,25,50,100]
)
plt.plot(width,test_test_scores[0])
plt.plot(width,test_test_scores[1])
plt.plot(width,test_test_scores[2])
plt.plot(width,test_test_scores[3])
plt.plot(width,test_test_scores[4])
plt.show()
```

The optimal model for test set accuracy is one with a learning rate of 0.01 and either 5 or 25 to 50 hidden units.
